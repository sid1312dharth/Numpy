```python
ran=np.random.randint(1,100,20).reshape(4,5)
```

```python
ran
```

```python
ran>50
```

```python
#indexing using Boolean array
ran[ran>50]
```

```python
ran[(ran>50) & (ran%2!=0)]
```

```python
ran[(ran>50) & (ran%2!=0)]=0
```
