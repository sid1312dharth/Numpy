```python
"""
Matrix Multiplication (Definition)

Matrix multiplication is a mathematical operation where two matrices are combined to produce a new matrix by multiplying rows of the first matrix with columns of the second matrix.

🧠 Simple Definition

Matrix multiplication is the process of taking the dot product of rows and columns from two matrices to form a new matrix.

📏 Condition (Important Rule)

If:

First matrix A has size (m × n)
Second matrix B has size (n × p)

👉 Then multiplication is possible, and the result will be a matrix of size (m × p).

🔢 How it works (Concept)

Each element of the result matrix is calculated as:

Take a row from matrix A
Take a column from matrix B
Multiply corresponding elements
Add them together
"""


import numpy as np
a=np.arange(2,6).reshape(2,2)
b=np.arange(8,12).reshape(2,2)
c=np.dot(a, b)
print(c)
```

```python
import numpy as np

# First matrix (2x2)
A = np.array([[1, 2],
              [3, 4]])

# Second matrix (2x2)
B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication
C = np.dot(A, B)

print(C)
```
